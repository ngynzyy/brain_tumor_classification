```python
def download_and_unzip(url, extract_to='.'):
    http_response = urlopen(url)
    zipfile = ZipFile(BytesIO(http_response.read()))
    unzip_name = os.path.basename(url)  # Use the filename as the folder name
    folder_path = os.path.join(extract_to, unzip_name)
    os.makedirs(folder_path, exist_ok=True)
    zipfile.extractall(path=folder_path)
```

```python
for i in downzip:
  try:
    download_and_unzip(i, extract_to='/content/mat')
  except Exception as e:
    print(f"Can't download from {i} because: {e}")
```

* Vì tất cả file đã đọc là file `.mat` nên ta phải trích xuất ma trận ảnh từ những file này cùng với những thông tin liên quan: `PatientID, Label, TumorBorder`.

```python
# Create 3 empty folder if not exists
def create_folder(path:str):
  if os.path.exists(path): # Exist = Delete
    shutil.rmtree(path)
  os.makedirs(path) # Create a new one

create_folder('png')

# Convert mat files in /mat to png files in /png
def convert_mat_to_png(mat_file_path, output_dir):
    data = mat73.loadmat(mat_file_path)

    imageArray = np.array(data['cjdata']['image'])
    label = int(data['cjdata']['label'])
    patientID = data['cjdata']['PID']
    vecBorder = np.array(data['cjdata']['tumorBorder'])

    hi = np.max(imageArray)
    lo = np.min(imageArray)
    imageArray = (((imageArray - lo)/(hi-lo))*255).astype(np.uint8)

    # Create the output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)
    image = Image.fromarray(imageArray)

    # Convert and save the data as a PNG file
    output_file_path = os.path.join(output_dir, os.path.splitext(os.path.basename(mat_file_path))[0] + '.png')
    image.save(output_file_path)

    return output_file_path, label, patientID, vecBorder

def convert_imgs(mat_folder):
    paths = []
    labels = []
    ID = []
    borders = []
    for root, dirs, files in os.walk(mat_folder):
        for file in files:
            if file.endswith('.mat'):
                mat_file_path = os.path.join(root, file)
                filepath, label, id, bord = convert_mat_to_png(mat_file_path, '/content/png')
                paths.append(filepath)
                labels.append(label)
                ID.append(id)
                borders.append(bord)
    return paths, labels, ID, borders
```

* Trong dataframe vừa tạo có `VecBorder` chứa tọa độ $(x_i,y_i)$ các điểm bao quanh khối u ở não. Dựa vào đó, ta sẽ dùng `Shoelace Formula` tính diện tích một đa giác khi biết tọa độ đỉnh tạo thành đa giác đó:
$$\text{Area} = \frac{1}{2} \left| \sum_{i=1}^{n} (x_iy_{i+1} - x_{i+1} y_i) \right|$$
> **Proof:** (by Green Theorem) https://math.stackexchange.com/a/1558415
* Một đỉnh tương ứng với một **"gai"** của khối u, ta cũng sẽ dựa vào đó để tìm số gai tương ứng để thực hiện về sau.


```python
def calculate_area(coordinates):
    n = len(coordinates) // 2  # Divide by 2 since each coordinate pair has two values (x, y)
    area = 0

    # Apply the Shoelace formula
    for i in range(n):
        x1, y1 = coordinates[2*i], coordinates[2*i + 1]
        x2, y2 = coordinates[(2*i + 2) % len(coordinates)], coordinates[(2*i + 3) % len(coordinates)]
        area += (x1 * y2 - x2 * y1)

    area = abs(area / 2)  # Take the absolute value and divide by 2
    return area
```

```python
df_new['AreaBorder'] = np.round((df_new['VecBorder'].apply(calculate_area)),2)
df_new['borderSpikes'] = df_new['VecBorder'].apply(len)
df_new['borderSpikes'] = df_new['borderSpikes'].apply(lambda x : x//2)
df_new
```

#### 2.2.2 Handling Outlier Values In Tabular Data


* Hàm `rm_outliers()` được sử dụng để loại bỏ giá trị ngoại lai trong dữ liệu.
* Công thức IQR được tính như sau:
$$IQR = Q3- Q1$$
Trong đó, Q1 là phân vị thứ 25% và Q3 là phân vị thứ 75% của các giá trị trong một mẫu dữ liệu.



```python
def rm_outliers(df: pd.DataFrame, column: str)->pd.DataFrame:
  a = np.array(df[column])
  q3, q1 = np.percentile(a, [75, 25])
  iqr = q3 - q1
  upper_bound = q3 + 1.5 * iqr
  lower_bound = q1 - 1.5 * iqr
  for i in range(len(df)):
    if (df[column][i] > upper_bound) or (df[column][i] < lower_bound):
      df = df.drop(index=[i])
  return df
```

* Tiếp theo, chúng em sử dùng hàm `rm_outliers` để loại bỏ các giá trị ngoại lai trong cột `AreaBorder` của bảng dữ liệu `df_reg` và gán lại bảng dữ liệu đã được lọc vào biến `df_reg`.

```python
df_reg = rm_outliers(df_reg, 'AreaBorder')
display(df_reg)
```

#### 2.2.3 Scaling and Normalization


* Với data tabular, chúng ta sẽ chuẩn hóa dữ liệu (`X`) theo Standard Scaler, bằng thư viện `sklearn`.
* Mỗi điểm dữ liệu sẽ được chuẩn hóa theo công thức sau:
$$
z = \frac{x - \mu}{s}
$$
Trong đó:
  * $z$ là giá trị sau khi được chuẩn hóa, $x$ là giá trị ban đầu.

#### 2.2.1 Linear Regression Overview


* Do sự tương quan của chúng, khi khối u có diện tích lớn thì số gai góc cũng lớn và ngược lại. Nhưng liệu còn yếu tố nào ảnh hưởng đến diện tích khối u ngoài số lượng gai (chẳng hạn như thời gian bệnh nhân phát bệnh, mức độ nghiêm trọng của khối u,...).

$$Y = \beta_0 + \beta_1X + \epsilon$$
Trong đó:

  * Y: là biến phụ thuộc (được giải thích bởi biến độc lập X).
  * X là biến độc lập (predictor variable) giải thích sự biến đổi của biến phụ thuộc.
  * $\beta_0$: là hệ số chặn (intercept).
  * $\beta_1$: là hệ số của biến độc lập (slope).
  * $\epsilon$: là số ngẫu nhiên (residual error) của mô hình, $\epsilon$ tuân theo phân phối chuẩn.
* Trong bài toán này, Y là `AreaBorder` và X là `borderSpikes`.

Mục tiêu của mô hình OLS là tìm giá trị $\hat{\beta}$ sao cho tổng bình phương sai số (residuals) giữa giá trị dự đoán $\beta X$ và giá trị thực tế $Y$ là nhỏ nhất.

```python
def crop_img(img, scale=0.85):
    height, width = img.shape
    center_x, center_y = width / 2, height / 2
    width_scaled, height_scaled = width * scale, height * scale
    left_x, right_x = center_x - width_scaled / 2, center_x + width_scaled / 2
    top_y, bottom_y = center_y - height_scaled / 2, center_y + height_scaled / 2
    img_cropped = img[int(top_y):int(bottom_y), int(left_x):int(right_x)]
    return img_cropped
```

#### 3.1.2 Feature Scaling (Normalization)

* Tiếp theo, ta dùng hàm `scale_img()` để chuẩn hóa các điểm ảnh nằm trong cùng một khoảng giá trị từ 0 đến 1, giúp mô hình tránh bias khi các khoảng chênh lệch giữa các điểm ảnh quá lớn.
* Chúng ta sẽ chuẩn hóa dữ liệu (x) theo `normalization` từ thư viện `sklearn` với công thức như sau:
$$y = \frac{x - min}{max - min}$$
  Trong đó:
  * $x$ là giá trị của một điểm dữ liệu.
  * $min$ là giá trị nhỏ nhất của tất cả các điểm dữ liệu.
  * $max$ là giá trị lớn nhất của tất cả các điểm dữ liệu.
  * $y$ là giá trị của điểm dữ liệu sau khi được chuẩn hóa. Các giá trị $y$ sẽ thuộc vào khoảng `[0, 1]`.


```python
def scale_img(img):
    # normalized_image_array = img / 255.0
    # Flatten the image array
    flattened_image = img.flatten()
    reshaped_image = flattened_image.reshape(-1, 1)

    # Use MinMaxScaler to scale images
    scaler = MinMaxScaler()
    scaled_image = scaler.fit_transform(reshaped_image)

    scaled_image_array = scaled_image.reshape(img.shape)
    scaled_image_array = np.clip(scaled_image_array, 0, 1)
    return scaled_image_array
```

#### 3.1.3 Dimension Reduction (to compress images)


* Bước tiếp theo, ta sử dụng PCA để nén ảnh / giảm kích thước ảnh bằng cách transform ma trận ảnh từ $n$ chiều thành $k$ chiều, tức giữ lại một số feature có ý nghĩa, lý tưởng nhưng vẫn không làm giảm chất lượng ảnh. Từ đó, ta có thể giảm kích thước ảnh đi rất nhiều lần và điều này cũng làm giảm sức tải cho model.

**Giả sử:** input là ma trận $A_{m \times n}$ (trong data của mình là $512 \times 512$).
1. Đầu tiên, ta cần chuẩn bị một datasets trên grayscale (thay vì là RGB) bằng cách dùng `image.convert("L")` trong mỗi pixel ảnh, ngoài ra cũng cần $normalize$ ảnh để mỗi điểm ảnh nằm trong khoảng $[0;1]$.
2. Tiếp theo, ta tính ma trận $Cov$ là ma trận $S$ bằng cách nhân $A = A_{normalized}$ với chuyển vị của chính nó, và tìm các `eigenvector` $u$, các `eigenvalue` $\lambda$ bằng cách giải phương trình đặc trưng $(*)$.

$$
S = A^T * A \\
S * u = \lambda * u \hspace{.5cm} (*)
$$

3. Sắp xếp các `eigenvector` $u$ theo `eigenvalue` $\lambda$ giảm dần.

4. Lấy $k$ `eigenvector` ứng với $k$ `eigenvalue` lớn nhất và xếp chúng thành ma trận $P_{n \times k}$.

5. Có được ma trận ảnh đã nén $B$:
$$
B_{m \times k} = A_{m \times n} * P_{n \times k}
$$

6. Reconstruct (tái tạo lại) ảnh theo shape cũ bằng cách nhân ma trận nén $B$ với ma trận chuyển vị của ma trận thành phần chính $P$ chuyển vị để tái tạo lại hình ảnh ban đầu $A'$:
$$
A'_{m \times n} = B_{m \times k} * P_{n \times k}^T
$$

> **Minh họa** cho việc reconstruct lại ảnh (trong ví dụ sẽ lấy 4 components (là đặc trưng của ảnh) để tái tạo lại bức ảnh, output sẽ là một ảnh mới với kích thước nhỏ hơn:
![](https://i.imgur.com/8Iy3oMA.png)

```python
def compress_img(image_scaled):
    pca = PCA()
    transformed_image = pca.inverse_transform(pca.fit_transform(image_scaled.reshape(-1, 1)))
    transformed_image_gray = transformed_image.reshape(image_scaled.shape)
    return transformed_image_gray
```

##### 3.2.1.a Calculate Entropy Value


Trước tiên, cần tính giá trị **Entropy** để đánh giá mức độ phân tán và đa dạng của dữ liệu. **Entropy** được tính theo công thức:

$$H(Y) = -\sum_{i=1}^k P(Y=y_i)\log_2P(Y=y_i)$$

```python
# Specify the path to the main folder containing subfolders
data_folder = 'pca'

# Initialize lists to store features and labels
images = []
labels = []


# Load data for handling decision tree
for label in os.listdir(data_folder):
    label_folder = os.path.join(data_folder, label)
    if not os.path.isdir(label_folder):
        continue  # Skip non-folder files

    progress_bar = tqdm(total=len(os.listdir(label_folder)), desc=f"Loading images from label {label}")

    for image_file in os.listdir(label_folder):
        image_path = os.path.join(label_folder, image_file)

        # Read the image from file
        image = cv2.imread(image_path)

        # Append image and corresponding label to the lists
        images.append(image)
        labels.append(int(label))  # Convert label to integer

        progress_bar.update(1)

    progress_bar.close()

def calculate_entropy(labels):
    unique_labels, counts = np.unique(labels, return_counts=True)
    probabilities = counts / len(labels)
    entropy = -np.sum(probabilities * np.log2(probabilities))
    return entropy

print(f"\nEntropy: {calculate_entropy(labels)}")
```

Chỉ số **Entropy** tính dựa trên biến phụ thuộc `label` của bộ dữ liệu này là $1.52403$. Trong khi đó, bộ dữ liệu có 3 labels khác nhau tương ứng với 3 loại khối u não, vì vậy giá trị **Entropy** nằm trong khoảng $0$ đến $log_23 = 1.58496$. Do đó, giá trị **Entropy** hiện tại là khá cao, cho thấy các mẫu dữ liệu không tập trung và phân phối không đồng đều (đã được thể hiện ở bước EDA). Điều này làm cho việc tìm ra các quy tắc phân chia và đưa ra quyết định chính xác của mô hình Decision Tree trở nên khó khăn.

##### 3.2.1.c Training the Decision Tree Model

Chúng ta sẽ xây dựng mô hình Decision Tree trên bộ data đã được tiền xử lý dựa trên công thức:

$$\text{Information Gain}(X) = H(Y) - H(Y|X)$$

Trong đó:
  - $\text{Information Gain}(X)$: lượng thông tin mới mà một thuộc tính cung cấp khi biết giá trị của thuộc tính đó.
  - $H(Y)$: giá trị **Entropy** của biến phụ thuộc $Y$.
  - $H(Y|X)$: **Entropy** của biến phụ thuộc $Y$ khi biết giá trị của thuộc tính X.

Ở bài toán này, $Y$ là `label` và $X$ là `feature from Image`

```python
# Split the dataset into training and testing sets
X_train_tree, X_test_tree, y_train_tree, y_test_tree = train_test_split(features, labels, test_size=0.2, random_state=42)

# Create a decision tree classifier
clf = DecisionTreeClassifier()

# Train the decision tree classifier
clf.fit(X_train_tree, y_train_tree)

# Make predictions on the test set
y_pred_tree = clf.predict(X_test_tree)
```

##### 3.2.2.b Model Architecture for CNN


* Cấu trúc CNN (customized) mà chúng ta sử dụng là:
  > (Convolutional Layer (Conv) $\rightarrow$ ReLU $\rightarrow$ Pooling layer (Pool))$\times 3$ -> Fully Connected layer (FC) + ReLU  $\rightarrow$ FC $\rightarrow$ Softmax
* **Conv2D (tích chập hai chiều):**

$$
(f*g)(x, y) = \int_{u,v} f(u,v) g(x-u,y-v) du dv = (g*f)(x, y)
$$
$$
(f*g)[x,y]=\sum_{i,j=-\infty}^\infty f[i,j]g[x-i,y-j]
$$
<center> <img src="https://zigzag-marlin-d4d.notion.site/image/https%3A%2F%2Fvitalflux.com%2Fwp-content%2Fuploads%2F2020%2F11%2Fcnn_convolve_with_padding.gif?id=86a1fbb7-05da-4230-81d9-5dcf85d91a8b&table=block&spaceId=7f10308b-9c45-4e5b-b9fa-69da441ff8c9&userId=&cache=v2" width="500"> </center>

* **ReLU (Activation Function):** như một hàm kích hoạt, trả về giá trị 0 nếu biến đầu vào là âm.
$$f(x) = max(0,x)$$
* **Pooling layer (Pool):** Thường được sử dụng sau lớp Convolutional làm đơn giản hóa thông tin, giảm số lượng tham số / tính toán ở các lớp sau. Tăng tính ổn định với các biến đổi nhỏ trong đầu vào.
<center> <img src="https://qph.fs.quoracdn.net/
main-qimg-81a0fa489ea440d54e72513f5a4c0b35" width="500"> </center>

* **Fully Connected Layer (FC):** kết nối các đặc trưng đã trích xuất từ các convolutional và pooling layer để phân loại hoặc dự đoán giá trị đầu ra của mô hình. Nó cung cấp sự liên kết giữa các đặc trưng để đưa ra một dự đoán chính xác hơn, tăng cường khả năng phân loại của mạng CNN.
<center> <img src="https://i.stack.imgur.com/YYCJI.png" width="500"> </center>

* **Softmax:** một activation function thường được sử dụng ở layer cuối của bài toán classification để phân loại đầu ra. Ở đó đầu ra sẽ là xác suất dự đoán rơi vào các class.
<center> <img src="https://images.viblo.asia/f14eef5e-1672-4ac7-b68b-cda2da0787d0.png" width="300"> </center>


```python
class BrainTumorCNN(nn.Module):
    def __init__(self):
        super(BrainTumorCNN, self).__init__()
        self.conv_layers = nn.Sequential(
            nn.Conv2d(1, 16, kernel_size=3, stride=1, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )
        self.fc_layers = nn.Sequential(
            nn.Linear(64 * 8 * 8, 128),
            nn.ReLU(inplace=True),
            nn.Linear(128, 3),  # 3 classes: glioma, meningioma, pituitary tumor
            nn.LogSoftmax(dim = 1)
        )

    def forward(self, x):
        x = self.conv_layers(x)
        x = x.view(x.size(0), -1)
        x = self.fc_layers(x)
        return x
```

##### 3.2.2.c Training the CNN Model


* Tiến hành train mô hình và đánh giá các thông số đi kèm mỗi lần thực thi(mỗi epoch) theo hai chỉ số `loss` và `accuracy`.
*Hàm Loss (hàm mất mát): tổng các lỗi mắc phải đối với từng mẫu trên tập train hoặc tập validation; xác định mức độ của kết quả dự đoán dựa trên so với giá trị thực cần dự đoán. Nếu mô hình dự đoán sai nhiều thì giá trị của loss function sẽ càng lớn và ngược lại.
* Có nhiều hàm loss, ở đây chúng ta sẽ sử dụng `Cross-entropy` được sử dụng để học phân bố xác suất của dữ liệu (tạo ra các mô hình với độ chắc chắn cao (precision, recall cao))
$$loss(x,y) = -\sum{x\log{y}}$$
với: x là xác suất label đúng, y là xác suất label dự đoán.
* Accuracy: tỉ lệ các trường hợp được dự đoán đúng trên tổng số các trường hợp.


```python
# Create an instance of the CNN model
model = BrainTumorCNN().to(device)

# Define the loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training loop
num_epochs = 25
train_loss_list, val_loss_list, accuracy_list = [], [], []

for epoch in range(num_epochs):
    train_loss = 0.0
    val_loss = 0.0
    correct = 0.0
    # Training
    model.train()
    loop = tqdm(enumerate(train_loader), total = len(train_loader), bar_format=f"Epoch [{epoch+1}".ljust(10) + f"/ {num_epochs}]\t" + '{bar}{r_bar}')
    for batch_idx, (images, labels) in loop:
        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        # Progress bar version kontrimnon
        loop.set_postfix(loss = loss.item())
        # Update train_loss
        train_loss += loss.item() * images.size(0)

    # Validation
    model.eval()
    with torch.no_grad():
        for images, labels in val_loader:
            images = images.to(device)
            labels = labels.to(device)

            outputs = model(images)
            loss = criterion(outputs, labels)
            val_loss += loss.item() * images.size(0)

            _, predicted = torch.max(outputs.data, 1)
            correct += (predicted == labels).sum().item()

    train_loss_list.append(train_loss / len(train_dataset))
    val_loss_list.append(val_loss / len(val_dataset))
    accuracy_list.append(correct / len(val_dataset))
    # Print the loss and the accuracy of kontrimnon
    print(f"Train Loss (final): {train_loss_list[-1]:.4f} | Val Loss: {val_loss_list[-1]:.4f} | Accuracy: {accuracy_list[-1] * 100:.4f}%")
```

  * $Precision = \frac{TP}{TP+FP}$ (trong tất cả dự đoán Positive, bao nhiêu dự đoán là chính xác?)
  * $Recall = \frac{TP}{TP+FN}$ (trong tất cả trường hợp Positive, bao nhiêu trường hợp dự đoán là chính xác?)
  $$F_1 = 2 \frac{Precision*Recall}{Precision+Recall}$$(kết hợp cả hai chỉ số trên), $F_1$ cao khi cả hai chỉ số đều cao, trường hợp tốt nhất $F_1$ gần bằng 1, xấu nhất gần bằng 0.

#### **3.2.3 Gradient Boosting**


* Mô hình CNN học được đặc trưng hiệu quả từ xử lý hình ảnh, trong khi đó Gradient Boosting sử dụng học có giám sát để huấn luyện mô hình để thực hiện dự đoán. Chúng ta có thể sử dụng các đặc trưng học được từ CNN để huấn luyện một mô hình Gradient Boosting như XGBoost và kết hợp cả hai để cải thiện độ chính xác của mô hình trên các tập dữ liệu mới.
* Công thức tổng quát cho mô hình XGBoost là:
$$T = \sum_{h}\lambda_hT_H$$
Trong đó:

 * $T_H$ là poor fit của model trên dataset.
 * $\lambda$ là learning rate của mô hình boost.
 * $T$ là mô hình sau khi được boost.

<img src = "https://uc-r.github.io/public/images/analytics/gbm/learning_rate_comparison.png" width = 800>

##### 3.2.3.a Training XGBoost Model

1. Khởi tạo predictions ban đầu từ mô hình trước đó:
   -  $\hat{y}_i^0 = \text{predictions}_i$ (dự đoán từ mô hình CNN).

2. Tính toán hàm mất mát và đạo hàm:
   - Hàm mất mát: $L(y_i, \hat{y}_i)$. Đây là đo lường sai số giữa predicted label $\hat{y}_i$ và true label $y_i$.
   - Gradient: $g_i = \frac{{\partial L(y_i, \hat{y}_i)}}{{\partial \hat{y}_i}}$. Đây là đạo hàm riêng của hàm mất mát theo label hiện tại.

3. Lặp qua các vòng lặp (iterations):

   a. Tính gradient:
      - Gradient: $g_i = \frac{{\partial L(y_i, \hat{y}_i^{t-1})}}{{\partial \hat{y}_i^{t-1}}}$.

   b. Xây dựng Decision Tree:
      - Sử dụng gradient tính toán từ bước trước, xây dựng một Decision Tree mới để giảm thiểu hàm mất mát.

   c. Cập nhật dự đoán của mô hình:
      - Cập nhật dự đoán: $\hat{y}_i^t = \hat{y}_i^{t-1} + \lambda \cdot \text{predictions}$.
      - Đây là sự kết hợp giữa dự đoán từ CNN và dự đoán mới từ cây quyết định, với $\lambda$ là learning rate (tốc độ học của mô hình).

4. Lặp lại các bước 3a-3c cho số lần lặp tối đa (`num_rounds`), hoặc gặp điều kiện dừng.

```python
test_features = []
model.eval()
with torch.no_grad():
    for images, _ in val_loader:
        images = images.to(device)
        features = model.conv_layers(images)
        features = features.view(features.size(0), -1)
        test_features.append(features.cpu().numpy())

test_features = np.concatenate(test_features, axis=0)

predictions = np.array(predictions)
true_labels = np.array(true_labels)

dnn_dmatrix = xgb.DMatrix(test_features, label=true_labels)

# Define XGBoost parameters
params = {
    'max_depth': 3,
    'eta': 0.01, # lambda - learning rate
    'objective': 'multi:softmax',  # Use 'multi:softmax' for multiclass classification
    'num_class': 3,  # 3 types of brain tumors
}

# Train the XGBoost model
num_rounds = 100
xgb_model = xgb.train(params, dnn_dmatrix, num_rounds)
xgb_pred = xgb_model.predict(dnn_dmatrix)
```
